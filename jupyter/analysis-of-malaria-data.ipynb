At enrolment, women underwent a standardized history and clinical exam and had blood collected by venipuncture for baseline laboratory testing, filter paper sample and storage for future use. Baseline laboratory testing consisted of a thick blood smear for malaria parasites (which was read later by laboratory technicians not involved in study participants care and was not used for clinical care), qPCR for malaria parasites, CBC (hemoglobin, white blood cell count neutrophil count, platelet count) and ALT. After enrolment, pregnant women were reviewed routinely every 4 weeks for study drug administration and all clinical and laboratory data was recorded onto standardized case record forms (CRFs) by study clinicians. Information collected every 4 weeks during pregnancy included clinical data from history and clinical exam and results from laboratory tests which included blood smears for determination of malaria parasites, and qPCR for malaria parasite. Routine CBC and ALT were done at 20, 28, and 36 weeks of gestation. At any visit (scheduled or unscheduled), pregnant women who presented with a history of fever in the past 24 hours or a temperature > 38.0ÂºC (tympanic) had an urgent thick blood smear read in the study clinic. Women with a positive urgent thick blood smear were diagnosed with malaria and treated according to Uganda ministry of health guidelines.

Baseline covariates: age, gestational age, gravidity, ITN (insecticide bednet) coverage, reported sleeping under an ITN the previous night, Haemoglobin concentration, and detection of malaria parasites, are balanced between treatment groups (Table 1 [paper link](https://urldefense.com/v3/__https://www.sciencedirect.com/science/article/pii/S0140673618322244?via*3Dihub__;JQ!!OToaGQ!rPpsR7wM592O42J0Ryz6AKAIoeoeVGiOsTsMDCS9FdUlLbicwn82TS8JiZiVsx0wXacgSum86-K7dRogOxKJzNOQZQ$)).

# Study results:

- DP treatment is more effective than the SP treatment in preventing malaria during pregnancy. [paper link](https://urldefense.com/v3/__https://www.sciencedirect.com/science/article/pii/S0140673618322244?via*3Dihub__;JQ!!OToaGQ!rPpsR7wM592O42J0Ryz6AKAIoeoeVGiOsTsMDCS9FdUlLbicwn82TS8JiZiVsx0wXacgSum86-K7dRogOxKJzNOQZQ$)

- Gravidity is an effect modifier: adverse birth outcomes were higher in primigravid women than in multigravid women. [paper link](https://urldefense.com/v3/__https://www.sciencedirect.com/science/article/pii/S0140673618322244?via*3Dihub__;JQ!!OToaGQ!rPpsR7wM592O42J0Ryz6AKAIoeoeVGiOsTsMDCS9FdUlLbicwn82TS8JiZiVsx0wXacgSum86-K7dRogOxKJzNOQZQ$)

- Education and income are associated with risk of malaria during pregnancy. [paper link](https://urldefense.com/v3/__https://www.ncbi.nlm.nih.gov/pmc/articles/PMC6480498/__;!!OToaGQ!rPpsR7wM592O42J0Ryz6AKAIoeoeVGiOsTsMDCS9FdUlLbicwn82TS8JiZiVsx0wXacgSum86-K7dRogOxKS02RkXQ$)

```python
# Load packages
import pandas as pd
```

```python
# Participant-level data file path
p_path = "../data/PROMO-1_Participant_subsettedData.txt"
# Load participant-level data: 782 observations
ps = pd.read_csv(p_path, delimiter='\t')
# Variables of interests: has bednet, has ITN (insecticide bednet), 
# existing medical conditions at enrollment: hypertension, diabetes, ..., alcohol use,
# education
print(ps.shape)
ps.columns
```

```python
# Household-level data file path
hh_path = "../data/PROMOTE_BC3_cohort_Households.txt"
# Load data
hh = pd.read_csv(hh_path, delimiter='\t', low_memory=False)
# Select for variables of interest: wealth index category, household insecticide-net, dwelling type
print(hh.shape)
hh.columns
hh = hh[['Household_Id', 'Household ITNs [EUPATH_0044136]', 'Household wealth index, categorical [EUPATH_0000143]',
        'Dwelling type [ENVO_01000744]']]
```

```python
# Participant-repeated-measurement-level data file path
prm_path = "../data/PROMO-1_Participant repeated measure_subsettedData.txt"
# Load data
prm = pd.read_csv(prm_path, delimiter='\t', low_memory=False)
# Variables of interests
print(prm.shape)
prm.columns
```

```python
# Merge data based on participant and household ID
hh_ps = pd.merge(hh, ps, left_on='Household_Id', right_on='Household_ID', how='left')
full_df = pd.merge(prm, hh_ps, on=['Participant_ID', 'Household_ID'], how='left')
full_df.head()
```
